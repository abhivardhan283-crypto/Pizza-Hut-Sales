```python
import pandas as pd
import numpy as np
```

```python
orders = pd.read_csv("orders.csv")
order_details = pd.read_csv("order_details.csv")
pizzas = pd.read_csv("pizzas.csv")
pizza_types = pd.read_csv("pizza_types.csv", encoding="latin1")

print(orders.shape)
print(order_details.shape)
print(pizzas.shape)
print(pizza_types.shape)
```

```python
orders.head()
order_details.head()
pizzas.head()
pizza_types.head()

orders.info()
order_details.info()
```

```python
orders.isnull().sum()
order_details.isnull().sum()
pizzas.isnull().sum()
pizza_types.isnull().sum()
```

```python
orders["date"] = pd.to_datetime(orders["date"], format="mixed", dayfirst=True)
orders["time"] = pd.to_datetime(orders["time"], format="%H:%M:%S", errors="coerce").dt.time
```

```python
orders["hour"] = pd.to_datetime(orders["time"].astype(str), format="%H:%M:%S").dt.hour
orders["day"] = orders["date"].dt.day_name()
orders["month"] = orders["date"].dt.month_name()
```

```python
df = order_details.merge(orders, on="order_id")
df = df.merge(pizzas, on="pizza_id")
df = df.merge(pizza_types, on="pizza_type_id")
```

```python
df["revenue"] = df["quantity"] * df["price"]
```

```python
df.describe()
df.isnull().sum()
```

```python
df.to_csv("pizza_sales_cleaned.csv", index=False)
```
